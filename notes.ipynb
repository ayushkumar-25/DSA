## Number Theory

### Find the occurrences of digit d in the range
[GeeksforGeeks](https://www.geeksforgeeks.org/dsa/find-the-occurrences-of-y-in-the-range-of-x/)

```python
# Brute force

def get_count(num, k):
    count = 0
    for i in range(1, num + 1):
        count += str(i).count(str(k))
    return count

print(get_count(25, 2))
```

## Slinding Window

### Print all subarrays: brute force

```python
# Using list slicing

def get_subarrays(arr):
    size = len(arr)
    subarrays = []

    for i in range(size):
        for j in range(i, size):
            subarrays.append(arr[i:j+1])
    return subarrays

res = get_subarrays([1, 2, 3, 4])
for e in res:
    print(e, end='\n')
```

```python
# Without using list slicing

def get_subarrays(arr):
    n = len(arr)
    subarrays = []

    for i in range(n):
        temp = []
        for j in range(i, n):
            temp.append(arr[j])
            subarrays.append(temp.copy())
    return subarrays

res = get_subarrays([1, 2, 3])
for e in res:
    print(e, end='\n')
```
