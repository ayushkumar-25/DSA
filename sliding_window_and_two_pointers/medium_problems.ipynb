## Longest Substring without Repeating Character
[LeetCode](https://leetcode.com/problems/longest-substring-without-repeating-characters/description/)

```python
def get_longest_substring(s: str) -> int:
    sett = set()
    left = 0
    max_len = 0

    for right in range(len(s)):
        # Check if the window is valid
        while s[right] in sett:
            sett.remove(s[left])
            left += 1

        # If window is valid
        sett.add(s[right])
        window_size = (right - left) + 1
        max_len = max(max_len, window_size)

    return max_len


# Test cases
print(get_longest_substring('abcabcbb'))
print(get_longest_substring('bbbbb'))
print(get_longest_substring('pwwkew'))
```

```python
# Print substring as well

def get_longest_substring(s: str):
    sett = set()      # current window chars
    left = 0          # window start
    max_len = 0       # best length
    start = 0         # best start index

    for right in range(len(s)):
        # shrink window if duplicate
        while s[right] in sett:
            sett.remove(s[left])
            left += 1

        sett.add(s[right])  # expand window

        window_size = right - left + 1

        # update best window
        if max_len < window_size:
            start = left
            max_len = window_size

    return max_len, s[start: start + max_len]


# Test cases
print(get_longest_substring('abcabcbb'))
print(get_longest_substring('bbbbb'))
print(get_longest_substring('pwwkew'))
```
