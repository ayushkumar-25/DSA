## Count all digits of a number
[GeeksforGeeks](https://www.geeksforgeeks.org/problems/count-total-digits-in-a-number/1)

```python
# Brute Force

num = abs(int(input('Enter number: ')))
count = 0

while num != 0:
  num = num // 10
  count += 1

print(count)
```

```python
# Logrithmic Method

import math

num = abs(int(input('Enter Number: ')))

count = int(math.log10(num)) + 1
print(count)
```

```python
# Pythonic Way

num = int(input('Enter Number: '))
len_count = len(str(abs(num)))
print(len_count)
```

## Reverse a number

[LeetCode](https://leetcode.com/problems/reverse-integer/description/)

```python
# Brute Force

num = int(input("Enter Number: "))
rev = 0

sign = -1 if num < 0 else 1
num = abs(num)

while num != 0:
  digit = num % 10
  rev = rev * 10 + digit
  num = num // 10

if rev <= -2**31 or rev >= 2**31 - 1:
  print(0)
else:
  print(rev * sign)
```

## Palindrome number
[LeetCode](https://leetcode.com/problems/palindrome-number/description/)

```python
# Brute Force

num = int(input("Enter Number: "))
dub = num
rev = 0

if num < 0:
  print(False)
elif num == 0:
  print(True)
else:
  while num != 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10
  if rev == dub:
    print(True)
  else:
    print(False)
```

```python
# Pythonic Way

num = int(input("Enter Number: "))
if str(num) == str(num)[::-1]:
  print(True)
else:
  print(False)
```

## Check if the number is Armstrong
[GeeksforGeeks](https://www.geeksforgeeks.org/problems/armstrong-numbers2727/1)

```python
# Brute Force

def is_armstrong(num: int) -> bool:
    if num < 0:
        return False

    k = len(str(num))
    total = 0
    n = num

    while n > 0:
        digit = n % 10
        total += digit ** k
        n //= 10

    return total == num


# Usage
number = 153
print(is_armstrong(number))
```
