## Frequency Counting / Frequency Hashing

**Hashing** is a technique that converts input data into a fixed-size value (hash) using a hash function, mainly used for fast data retrieval and secure storage.

**Hash function** is a function that takes input data and converts it into a fixed-size value (hash code), typically used for fast lookup and data integrity.

```python
# Integer counter using list (array)

arr = list(map(int, input().split()))
size = max(arr) + 1
hash_arr = [0] * size

# pre-compute
for x in arr:
    hash_arr[x] += 1

# fetch counts
q = int(input())
for _ in range(q):
    num = int(input())
    print(hash_arr[num])
```

```python
# Char counter using list (array)

arr = input().split()
size = ord(max(arr)) + 1
hash_arr = [0] * size

# pre-compute
for i in arr:
    hash_arr[ord(i)] += 1

# fetch counts
q = int(input())
for _ in range(q):
    char = input().strip()
    print(hash_arr[ord(char)])
```

```python
# Interger counter using dict (hashmap): if-else

arr = list(map(int, input().split()))
hash_map = dict()

# Pre compute
for i in arr:
    if i in hash_map:
        hash_map[i] += 1
    else:
        hash_map[i] = 1

# Fetch count
q = int(input())
for _ in range(q):
    num = int(input())
    print(hash_map[num])
```

```python
# Char counter using dict (hashmap): get method

arr = input().split()
hash_map = dict()

# pre-compute
for x in arr:
    hash_map[x] = hash_map.get(x, 0) + 1

# fetch count
q = int(input())
for _ in range(q):
    char = input().strip()
    print(hash_map[char])
```

## Count Elements With Maximum Frequency
[LeetCode](https://leetcode.com/problems/count-elements-with-maximum-frequency/description/)

```python
# Using HashMap: 3 passes

def max_frequency(nums):
    hash_map = dict()

    # pre-compute
    for x in nums:
        hash_map[x] = hash_map.get(x, 0) + 1  # pass 1

    # frequency counter
    max_val = max(hash_map.values())  # pass 2
    result = sum([v for v in hash_map.values() if max_val == v])  # pass 3

    return result

# Testcase
print(max_frequency([1,2,2,3,1,4]))
print(max_frequency([1,2,3,4,5]))
```

```python
# Using hash_arr: one-pass

def max_frequency(nums):
    hash_arr = [0] * 101
    max = 0
    res = 0

    # one-pass
    for x in nums:
        hash_arr[x] += 1
        f = hash_arr[x]
        if f > max:
            max = f
            res = f
        elif f == max:
            res += f

    return res

# Testcase
print(max_frequency([1,2,2,3,1,4]))
print(max_frequency([1,2,3,4,5]))
```
