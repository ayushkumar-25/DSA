## Frequency Counting / Frequency Hashing

```python
# Integer counter using list (array)

arr = list(map(int, input().split()))
size = max(arr) + 1
hash_arr = [0] * size

# pre-compute
for x in arr:
    hash_arr[x] += 1

# fetch counts
q = int(input())
for _ in range(q):
    num = int(input())
    print(hash_arr[num])
```

```python
# Char counter using list (array)

arr = input().split()
size = ord(max(arr)) + 1
hash_arr = [0] * size

# pre-compute
for i in arr:
    hash_arr[ord(i)] += 1

# fetch counts
q = int(input())
for _ in range(q):
    char = input().strip()
    print(hash_arr[ord(char)])
```

```python
# Interger counter using dict (hashmap): if-else

arr = list(map(int, input().split()))
hash_map = dict()

# Pre compute
for i in arr:
    if i in hash_map:
        hash_map[i] += 1
    else:
        hash_map[i] = 1

# Fetch count
q = int(input())
for _ in range(q):
    num = int(input())
    print(hash_map[num])
```

```python
# Char counter using dict (hashmap): get method

arr = input().split()
hash_map = dict()

# pre-compute
for x in arr:
    hash_map[x] = hash_map.get(x, 0) + 1

# fetch count
q = int(input())
for _ in range(q):
    char = input().strip()
    print(hash_map[char])
```
