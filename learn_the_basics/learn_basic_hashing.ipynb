## Frequency Counting / Frequency Hashing

```python
arr = list(map(int, input().split()))
size = max(arr) + 1
hash_arr = [0] * size

# Pre-compute
for x in arr:
    hash_arr[x] += 1

# Fetch counts
q = int(input())
for _ in range(q):
    num = int(input())
    print(hash_arr[num])
```

```python
arr = list(map(str, input().split()))
size = ord(max(arr)) + 1
hash_arr = [0] * size

# pre compute
for i in arr:
    hash_arr[ord(i)] += 1

print(hash_arr)

# Fetch counts
q = int(input())
for _ in range(q):
    char = input()
    print(hash_arr[ord(char)])
```
